###KNN

```
from sklearn.neighbors import KNeighborsClassifier as knn
```

####k-fold cross-validation

```
from sklearn.model_selection import cross_val_score as cvs
```

```
def val_score(k):
  mean=[]
  std=[]
  for i in range(1,k):
    score=cvs(knn(n_neighbors=i),X=x,y=y,cv=10)  ##return list of scores
    mean.append(score.mean())
    std.append(score.std())
  return mean,std
```

```
mean,std=val_score(50)
```

```
plt.plot(range(1,50),mean,color="green",label="mean")
plt.xlabel('k')
plt.ylabel('mean_score')
```

```
plt.plot(range(15,26),mean[14:25],color="green",label="mean")
plt.xlabel('k')
plt.ylabel('mean_score')
```

```
plt.plot(range(1,50),std,color="red",label="std")
plt.xlabel('k')
plt.ylabel('std_score')
```

```
plt.plot(range(15,26),std[14:25],color="red",label="std")
plt.xlabel('k')
plt.ylabel('std_score')
```

###Neural Network

```
%tensorflow_version 2.x  
import tensorflow as tf
from tensorflow import keras
```

```
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=[46]), #input layer takes each column as input hence,9
    keras.layers.Dense(64, activation='relu'),                  #hidden layer
    keras.layers.Dense(1,activation="sigmoid")                                       #output layer
  ])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```
model.summary()
```

####training

```
x_train_ann=df.drop(['capital_gain','capital_loss','salary'],axis=1)
x_train_ann=x_train.astype({"education_num":'int64'})

x_test_ann=test_df.drop(['capital_gain','capital_loss','salary'],axis=1)
x_test_ann=x_test_ann.astype({"education_num":'int64'})

ann=model.fit(x_train_ann,y,epochs=100)
print("train score ",model.evaluate(x_train_ann,y))
print("test score ",model.evaluate(x_test_ann,y_test))
```

```
pred_train_ann=model.predict_classes(x_train_ann)
print("confustion_matrix= \n",confusion_matrix(pred_train_ann,y))
print('classification_report= \n',classification_report(pred_train_ann,y))
```

####testing

```
pred_test_ann=model.predict_classes(x_test_ann)
print("confustion_matrix= \n",confusion_matrix(pred_test_ann,y_test))
print('classification_report= \n',classification_report(pred_test_ann,y_test))
```
